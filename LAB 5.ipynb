```python
import heapq
class WeightedGraph:
    def __init__(self):
        self.adj_list = {}

    def add_nodes(self, vertex):
        if vertex not in self.adj_list:
            self.adj_list[vertex] = []

    def add_edge(self, vertex1, vertex2, weight):
        if vertex1 in self.adj_list and vertex2 in self.adj_list:
            self.adj_list[vertex1].append((vertex2, weight))
            self.adj_list[vertex2].append((vertex1, weight))

    def get_neighbors(self, vertex):
        if vertex in self.adj_list:
            return self.adj_list[vertex]
        else:
            return []

    def get_vertices(self):
        return list(self.adj_list.keys())


    def prim(graph):
        # Initialize the algorithm with an arbitrary starting vertex
        start_vertex = graph.get_vertices()[0]
        visited = set([start_vertex])
        edges = graph.get_neighbors(start_vertex)
        heapq.heapify(edges)
        minimum_spanning_tree = []

        # Repeat until all vertices are included in the tree
        while len(visited) < len(graph.get_vertices()):
            # Get the minimum weight edge that connects a visited vertex to an unvisited vertex
            min_edge = None
            while edges:
                edge = heapq.heappop(edges)
                if edge[0] not in visited:
                    min_edge = edge
                    break
            if not min_edge:
                # If there are no more edges connecting visited and unvisited vertices, the graph is not connected
                return None

            # Add the minimum weight edge to the minimum spanning tree and update the visited set and the edges heap
            minimum_spanning_tree.append(min_edge)
            visited.add(min_edge[0])
            for neighbor_edge in graph.get_neighbors(min_edge[0]):
                if neighbor_edge[0] not in visited:
                    heapq.heappush(edges, neighbor_edge)

        return minimum_spanning_tree
    # KRUKSHUAL ALGORITHM
    def kruskal(graph):
        # Initialize the algorithm with a disjoint set for each vertex
        disjoint_set = DisjointSet(graph.get_vertices())
        edges = []
        minimum_spanning_tree = []

        # Add all edges to the heap
        for vertex in graph.get_vertices():
            for neighbor, weight in graph.get_neighbors(vertex):
                edges.append((weight, vertex, neighbor))
        heapq.heapify(edges)

        # Repeat until all vertices are in a single set
        while len(minimum_spanning_tree) < len(graph.get_vertices()) - 1:
            # Get the minimum weight edge that connects two disjoint sets
            weight, vertex1, vertex2 = heapq.heappop(edges)
            if disjoint_set.find(vertex1) != disjoint_set.find(vertex2):
                minimum_spanning_tree.append((vertex1, vertex2, weight))
                disjoint_set.union(vertex1, vertex2)

        return minimum_spanning_tree
    def remove_edge(tree, edge):
        tree.remove(edge)
    def all_spanningTree(graph):
        # Initialize the algorithm with an arbitrary starting vertex
        start_vertex = graph.get_vertices()[0]
        visited = set([start_vertex])
        edges = graph.get_neighbors(start_vertex)
        heapq.heapify(edges)
        spanningtree = []

        # Repeat until all vertices are included in the tree
        while len(visited) < len(graph.get_vertices()):
            # Get the minimum weight edge that connects a visited vertex to an unvisited vertex
            min_edge = None
            while edges:
                edge = heapq.heappop(edges)
                if edge[0] not in visited:
                    spanningtree.append(edge)
                    visited.add(edge[0])
                    for neighbor_edge in graph.get_neighbors(edge[0]):
                        if neighbor_edge[0] not in visited:
                            heapq.heappush(edges, neighbor_edge)
                return spanningtree
                    
```
